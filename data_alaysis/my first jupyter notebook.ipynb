# Exploratry data analysis

# analysis via Gender

```python
ax=sns.countplot(x='Gender',data=df)# for showing graphical of you gen data using data which is stored in df obj

for bars in ax.containers:# for labeling in graph 
    ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)#data grpd by gen qery like sql
sns.barplot(x='Gender',y='Amount',data=sales_gen)#asigning the column plce
```

*From the above graph we can see that most of the buyers are females and even their purchasing powr is more than males *

```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender')#hue means data order or acording to 'specific'

for bars in ax.containers:
    ax.bar_label(bars)
```

# analysis via Age

```python
sales_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

From the above graph we can see that most of the buyers are age group between 26-35 yrs females

# analysis via states

```python
sales_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(25,10)})# to set size of graph

sns.barplot(data=sales_state,x='State',y='Orders')
```

```python
sales_state_amt=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(25,10)})# to set size of graph

sns.barplot(data=sales_state_amt,x='State',y='Amount')
```

From the above graph we can see that most of the buyers are from state u.p, maharastra and karnatka
